```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_node("fact_trans")
```

```python
G.add_node("dim_customer")
G.add_node("dim_account")
G.add_node("dim_branch")
G.add_node("dim_state")
```

```python
G.add_node("dim_custid")
G.add_node("dim_account_type")
G.add_node("dim_branch_city")
G.add_node("dim_country")
```

```python
G.add_edge("fact_trans","dim_customer")
G.add_edge("fact_trans","dim_account")
G.add_edge("fact_trans","dim_branch")
G.add_edge("fact_trans","dim_state")
```

```python
G.add_edge("dim_customer","dim_custid")
G.add_edge("dim_account","dim_account_type")
G.add_edge("dim_branch","dim_branch_city")
G.add_edge("dim_state","dim_country")
```

```python
pos = nx.spring_layout(G,seed=42)
plt.figure(figsize=(10,7))
nx.draw(G,pos,with_labels=True,node_size=3000,font_size=9)
plt.title("Snowflake Schema Representation")
plt.show()
```
